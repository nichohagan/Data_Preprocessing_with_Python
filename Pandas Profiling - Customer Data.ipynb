### Data Preprocessing after the Profile Report

```python
# There are inconsistencies in the dataset which need cleansing 
```

```python
# Fill the NaN with empyty string
lista = lista.fillna('')
lista
```

```python
# Clean the "ssn" colunm; and add a new colunm "SSN"

import re                #We first import Regular Expression
def clean_number(s):
    s = re.sub('[^0-9]', '', s)      #replaced all non-digit characters with a blank '' 
    return s                         
lista["SSN"] = lista["ssn"].apply(clean_number)  #Adding a new column that will take the result of "column_number" as values
lista

#The re. sub() function is used to replace occurrences of a particular sub-string with 
#another sub-string. This function takes as input the following: The sub-string to replace.

#The [^0-9] expression is used to find any character that is NOT a digit.  
#Tip: Use the [0-9] expression to find any character between the brackets that is a digit.
```
